```python
data_with_english_sentiment_aspect_copy = pd.read_csv(r'drive/My Drive/Colab Notebooks/new/data_with_english_sentiment_aspect_copy.csv')
```

```python
data_with_english_sentiment_aspect_copy = data_with_english_sentiment_aspect_copy.drop(['Sentiment Score', 'Sentiment'], axis=1)
```

```python
data_with_english_sentiment_aspect_copy
```

```python
reviews = data_with_english_sentiment_aspect_copy["Review Text"]
```

```python
reviews.iloc[0]
```

```python
pol = []
for i in range(len(reviews)):
    pol.append(sentiment_scores(reviews[i]))
```

```python
pol
```

```python
labels = []
for i in range(len(pol)):
  if pol[i]['neg'] > pol[i]['pos']: #and pol[i]['neg'] > pol[i]['neu']:
    labels.append('neg')
  elif pol[i]['pos'] > pol[i]['neg'] :#and pol[i]['pos'] > pol[i]['neu']:
    labels.append('pos')
  else:
    labels.append('neu')
```

```python
data_with_english_sentiment_aspect_copy['Sentiments'] = labels
```

```python
data_with_english_sentiment_aspect_copy
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder().fit(data_with_english_sentiment_aspect_copy['Sentiment'])
labels = le.transform(data_with_english_sentiment_aspect_copy['Sentiment'])
```

```python
data_with_english_sentiment_aspect_copy['Sentiment'] = labels
```
