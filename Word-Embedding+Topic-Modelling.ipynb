```python
def data_preprocessor(text):
    tokens = word_tokenize(text)
    tokens = [w.lower() for w in tokens]
    tokens_clean = str.maketrans('', '', string.punctuation)
    tokens_cleaned = [w.translate(tokens_clean) for w in tokens]
    clean_words = [word for word in tokens_cleaned if word.isalpha()]
    final_review = " ".join(clean_words)
    tokens = word_tokenize(final_review)
    # words = [word for word in tokens if not word in stop_words]
    # words = [spell.correction(word) for word in words]
    words = [spell.correction(word) for word in tokens]
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

```python
# tokenize using gensim
def sent_to_words(sentences, deacc=True):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence)))
```

```python
def make_bigrams(words):
    bigram = gensim.models.Phrases(sentences=words, min_count=15)
    bigram_mod = gensim.models.phrases.Phraser(bigram)                # building bigram model
    return bigram_mod
```

```python
# genrating sentence vectors
def document_vector(word2vec_model, doc):
    # remove out-of-vocabulary words
    doc = [word for word in doc if word in word2vec_model.wv]
    return np.mean(word2vec_model.wv[doc], axis=0)
```

```python
# genrating sentence vectors for topics/aspects too
loc_vec=np.mean(model.wv[l], axis=0)
food_vec=np.mean(model.wv[f], axis=0)
staff_vec=np.mean(model.wv[s], axis=0)
exp_vec=np.mean(model.wv[e], axis=0)
service_vec=np.mean(model.wv[sv], axis=0)
room_vec=np.mean(model.wv[r], axis=0)
```

```python
# finding cosine similarty between sentence vectors and aspect vectors
from scipy import spatial

# similarity results to store
result_loc=[]
result_food=[]
result_staff=[]
result_exp=[]
result_sv=[]
result_room=[]

# iterating over the list of sentence vectors
for i in x:    
    result_loc.append(1 - spatial.distance.cosine(i, loc_vec))
    result_food.append(1 - spatial.distance.cosine(i, food_vec))
    result_staff.append(1 - spatial.distance.cosine(i, staff_vec))
    result_exp.append(1 - spatial.distance.cosine(i, exp_vec))
    result_sv.append(1 - spatial.distance.cosine(i, service_vec))
    result_room.append(1 - spatial.distance.cosine(i, room_vec))
```

```python
# genrating a new dataframe for such columns to store similairty score
column_names = ["Location", 'Food', 'Staff', 'Experience/Value', 'Service', 'Room']
similarityMatrix = pd.DataFrame(columns = column_names)
```

```python
similarityMatrix['Location'] = result_loc
similarityMatrix['Food'] = result_food
similarityMatrix['Staff'] = result_staff
similarityMatrix['Experience/Value'] = result_exp
similarityMatrix['Service'] = result_sv
similarityMatrix['Room'] = result_room
```
