```python
def removePunctuation(str):
    str = fix_text(str)
    # The maketrans() method returns a mapping table that can be used with the translate() method to replace specified characters
    str = str.translate(str.maketrans('','',string.punctuation))
    if not str:
        str = " "
    return str
```

```python
def checkIfEnglish(str):
    newstr = removePunctuation(str)
    if newstr != " ":
        if dictionary.check(newstr):
            return 1
    return 0
```

```python
# dropping non english rows and taking only rows which are in english 
list_to_drop = []
for index, row in data_with_english.iterrows():
    eachTopic = row["Review Title"].split()
    eachRow = row["Review Text"].split()          # taking review text here
    try:
        if not(checkIfEnglish(eachTopic[0]) +
               checkIfEnglish(eachRow[0]) +
               checkIfEnglish(eachRow[1]) +
               checkIfEnglish(eachRow[int(len(eachRow)/2)]) +
               checkIfEnglish(eachRow[-1])) >=4:
            eachList = [row["Property Name"], row["Review Rating"], row["Review Title"], row["Review Text"], row["Location Of The Reviewer"], row["Date Of Review"]]
            list_to_drop.append(eachList)
            data_with_english.drop(index, inplace=True)
    except Exception:
        data_with_english.drop(index, inplace=True)
        pass
```

```python
def data_preprocessor(text):
    tokens = word_tokenize(text)
    tokens = [w.lower() for w in tokens]
    tokens_clean = str.maketrans('', '', string.punctuation)
    tokens_cleaned = [w.translate(tokens_clean) for w in tokens]
    clean_words = [word for word in tokens_cleaned if word.isalpha()]
    final_review = " ".join(clean_words)
    tokens = word_tokenize(final_review)
    # words = [word for word in tokens if not word in stop_words]
    # words = [spell.correction(word) for word in words]
    words = [spell.correction(word) for word in tokens]
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

```python
review_text_sentence = []
review_title_sentence = []
```

```python
start_time = time.time()
for index, row in data_with_english_sentence_reviews.iterrows():
    processed_title_sentence = data_preprocessor(row['Review Title'])
    review_title_sentence.append(processed_title_sentence)
    processed_text_sentence = data_preprocessor(row['Review Text'])
    review_text_sentence.append(processed_text_sentence)
    if(index % 1000 == 0):
        print("processed %d reviews" % (index+1))
print('%.2f sec' % (time.time()-start_time))
```
